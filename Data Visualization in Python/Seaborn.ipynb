- Scatter plots are used to showcase the relationship amongst two continuous variables. eg: age vs weight, miles per gallon vs speed etc.

##### Distribution Plots
- Distribution plots basically showcase the distribution of data points pertaining to a continuous variable.
- There are 3 main types of distribution plots i.e. **Rug Plot, Histogram, KDE Plot**
- In a Rug Plot, a dash or tick line gets added for every single data point & here y-axis bears no significance.
- In a Histogram, we create a grouping in terms of range of values of the data called bins & then count the no of data points falling under each bin & plot them as bars.
- We can also add a KDE(Kernel Density Estimation) plot which will overlap the histogram.
- KDE plot basically represents the estimation of continuous probability curve for finite data samples.

```python
# let us first construct the rug plot.
plt.figure(figsize=(10, 5), dpi=100)
sns.rugplot(x="salary", data=df)
```

```python
# we can adjust the height of the ticks created for each data point in the rug plot.
plt.figure(figsize=(10, 5), dpi=100)
sns.rugplot(x="salary", data=df, height=0.5) # here 50% of the y-axis is taken in terms of height of the ticks.
```

```python
# let us now create the histogram.
sns.displot(data=df, x="salary")
```

```python
# we can also define the number of bins required in the histogram & also to set the background.
sns.set(style="dark") # now style is set for the rest of the visulaizations to come, in order to rest we need to restart the kernel of our notebook. 
sns.displot(data=df, x="salary", bins=10)
```

```python
# let us change the color of our bins & its edges inside the histogram.
sns.displot(data=df, x="salary", bins=10, color="purple", edgecolor="red", linewidth=3)
```

```python
# let us now try to overlap the KDE plot & rug plot on top of the histogram.
sns.displot(data=df, x="salary", kde=True, rug=True)
```

```python
# another alternative way of creating a histogram.
sns.histplot(data=df, x="salary")
```

```python
# in case we only require the KDE plot.
sns.kdeplot(data=df, x="salary")
```

```python
# in order to limit the range of our KDE plot.

# let us first create a random data set lets say on weight of 300 people.
np.random.seed(10)
weight_data = np.random.randint(0, 100, 300)
weight_df = pd.DataFrame(weight_data, columns=["Weight"])
weight_df
```

```python
# let us first create our KDE plot.
sns.kdeplot(data=weight_df, x="Weight") # observe that the lower limit has gne beyond zero & upper limit going beyond 100.
```

```python
# let us restrict the range of our KDE plot to 0 towards the lower limit & 100 towards the upper limit.
sns.kdeplot(data=weight_df, x="Weight", clip=[0, 100])
```

```python
# in order to adjust the bandwidth of the KDE plot in order to factor in more variation. Be careful while using this..!!
sns.kdeplot(data=weight_df, x="Weight", clip=[0, 100], bw_adjust=0.5) # as the bandwidth increases, the KDE plot will move towards normalization, default bw=1.0.
```

```python
# let us now add the shaded area below the KDE plot.
sns.kdeplot(data=weight_df, x="Weight", clip=[0, 100], shade=True)
```

##### Categorical Plots

###### **First Classification Type**
- Categorical Plots are basically the visulaization equivalent of the Pandas groupby() method call.
- Here we are showcasing aggregation in terms of various statistical measures basis any chosen categorical data column from a data frame, eg: total revenue per country, order count per product category.
- There are two main categorical plot types i.e. **Count Plot & Bar Plot**.
- The **Count Plot** showcases the count per category & the **Bar Plot** showcases a chosen metric per category.

###### **Second Classification Type**
- Apart from above two plots we can also see the distribution of a continuous data variable across multiple categories.
- We have four main categorical plots for creating distributions across multiple categories i.e. **Box Plot, Violin Plot, Swarm Plot, BoxenPlot**.
- The **Box Plot** showcases the distribution of a continuous variable per category by showcasing the same through quartiles.
- The data in the box plot is divided into 4 equal number of data points i.e. *first quartile/25th percentile, median/50th percentile, third quartile/75th percentile*.
- The **Violin Plot** plays a similar role as a box plot however it showcases the probability density via KDE & it looks like a mirrored KDE plot & resultant shape resembles a violin, hence the name.
- The **Swarm Plot** showcases all the individual data points in the distribution, however for large data sets it won't show all the data points rather will showcase the distribution for them.
- The **Boxen Plot** also known as Letter-Value plot is basically an expansion made upon the normal box plot.
-  The boxen plot automatically calculates multiple *quantiles*(not *quartile*) basis the number of data points in the distribution.

##### Comparison Plots
- These are basically a two dimensional version of the plots that we have seen so far.
- There are two main types of comparison plots i.e. **Joint Plot & Pair PLot**.
- In a *Joint Plot* we can map a histogram to each of the two variables along with their scatter plot & we can also adjust the scatterplot to be a hexagonal plot or a 2D KDE plot.
- A *Pair Plot* creates a histogram for each numerical data column & a scatterplot comparison amongst all these numerical data columns of a data frame.
- It is good to first create a subset of data based on the required columns before creating a pair plot & we can also draw a KDE plot instead of histogram.

```python
# let us first import a data set.
df = pd.read_csv("/Users/rahul_arora/Documents/StudentsPerformance.csv")
df.head()
```

```python
# let us create a joint plot between math score & reading score.
sns.jointplot(data=df, y="reading score", x="math score")
```

```python
# let us know change the scatter plot to a hexagonal plot.
sns.jointplot(data=df, y="reading score", x="math score", kind="hex")  # darker hex denotes densly populated data points.
```

```python
# alternaticely we can also adjust the alpha parameter in order to see the spread with scatter plot.
sns.jointplot(data=df, y="reading score", x="math score", alpha=0.3)
```

```python
# we can also create histogram blocks instead of scatter plot which will aligned to the outer histogram plots.
sns.jointplot(data=df, y="reading score", x="math score", kind="hist")
```

```python
# let us now replace the outer histogram plots & inner scatter plot with KDE plot.
sns.jointplot(data=df, y="reading score", x="math score", kind="kde")
```

```python
# we can also shade the inner KDE plot.
sns.jointplot(data=df, y="reading score", x="math score", kind="kde", shade=True)
```

```python
# let us now see the scatter plot divided by gender.
sns.jointplot(data=df, y="reading score", x="math score", hue="gender")
```

```python
# let us now create the pair plot between all the scores data columns of the data frame.
sns.pairplot(df)
```

```python
# let us now modify the above plot to see the view bifurcated by gender.
sns.pairplot(data=df, hue="gender")
```

```python
# we can also replace the kde plots with histograms.
sns.pairplot(data=df, hue="gender", diag_kind="hist")
```

```python
# notice that there are duplicate plots due to repeating columns, in order to show only unique columns plots.
sns.pairplot(data=df, hue="gender", corner=True)
```

##### Grid Plots
- Grid Plots automatically creates a grid based off a categorical column & it leverages matplotlib subplots() to do so.
- So as with the subplots where we have to explicitly pass the specific number of rows or rows, here we have to simply pass the column name & seaborn will automatically create the subplot grid. 

```python
# let us first create a box plot for writing score bifurcated by gender & further sub divided by whether the test preparation course was completed or not.
sns.catplot(data=df, y="math score", x="gender", row="test preparation course", kind="box")
```

```python
# we can also bifurcate the data basis test preparation couse category in columns rather than rows.
sns.catplot(data=df, y="math score", x="gender", col="test preparation course", kind="box")
```

```python
# we can also add in another dimension let's say race/ethnicity for further sub division of the data.
sns.catplot(data=df, y="math score", x="gender", row="test preparation course", col="race/ethnicity", kind="box")
```

```python
# we can also create other plots & explore them basis the categories defined.
sns.catplot(data=df, y="math score", x="gender", col="test preparation course", kind="violin")
```

##### Matrix Plots
- It is basically a visual representation of all the numeric data columns of a data frame & is the visual equivalent of a Pivot Table.
- There are two types of matrix plots i.e. **Heat Map & Cluster Map**.
- A heat map showcases the distribution of data values i.e. cells with a color mapping.
- In a *heat map* it's important to note that all the cells should have the same unit. 
- A *cluster map* first reorganizes a data into similar groups & then performs the color mapping.

```python
# let us first import a new data set.
world_df = pd.read_csv("/Users/rahul_arora/Documents/world_population_prospects.csv")
world_df
```

```python
# in order to create the heatmap for the above data we need to first set the countries as the index.
world_df = world_df.set_index("Countries")
world_df
```

```python
# now let us create the heatmap, one thing to be careful that we need to remove the life expectancy column as it is different unit compared to other columns.
sns.heatmap(world_df.drop("Life expectancy", axis=1))
```

```python
# let us add some spacing between the cells of the heat map so as to have a more clearer view.
sns.heatmap(world_df.drop("Life expectancy", axis=1), linewidths=0.1)
```

```python
# in order to display the cell values.
sns.heatmap(world_df.drop("Life expectancy", axis=1), linewidths=0.1, annot=True)
```

```python
# we can also change the color scheme of the heat map, viridis being one of the most popular one.
sns.heatmap(world_df.drop("Life expectancy", axis=1), linewidths=0.1, annot=True, cmap="viridis")
```

```python
# we can change the color scale of the heatmap mapping basis the defined center value which will then color the heat map accordingly.
sns.heatmap(world_df.drop("Life expectancy", axis=1), linewidths=0.1, annot=True, cmap="viridis", center=30)
```

```python
# let us now create a cluster map basis the data defined in the data frame.
sns.clustermap(world_df.drop("Life expectancy", axis=1), linewidths=0.1, annot=True, cmap="viridis")
```

```python
# in order to showcase only the clustering of regions.
sns.clustermap(world_df.drop("Life expectancy", axis=1), linewidths=0.1, annot=True, cmap="viridis", col_cluster=False)
```
