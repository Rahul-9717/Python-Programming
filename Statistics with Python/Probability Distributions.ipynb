#### **Probability Distributions**
- These are the distribution of probabilities of different events while carrying n number of trials in an experiment. 
- There are three basic probability distributions i.e. **Binomial, Poisson & Normal Dustributions**.

##### Binomial Distribution
- A binomial distribution is a discrete distribution that basically measures the number of successes.
- A binomial experiment consists of n repeated trials & each trial can result in just two possible outcomes i.e one is success & other, a failure.
- The probability of success, denoted by p, is the same on every trial.
- The trials are independent i.e. the outcome of one trial does not affect the outcome on other trials.
- Binomial Probability P(x) is the probability that an n-trial experiment results in exactly x successes, when the probability of success of an individual trial is p.
- Formula : **P(x) = (n! / x!(n-x)!) . p^x . (1-p)^n-x**
- Eg: If you flip a coin 4 times & we want to calculate the probability of getting 1 head, thus in this case n=4, x=1, p=0.50.
- Mean of the distribution = **n.p**, Variance = **np(1-p)** & Standard Deviation = **sqrt(np(1-p))**

```python
# let us first import the relevant libraries.
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt
import seaborn as sns

# let us now simulate the coin flipping trials.
# we are flipping 10 coins repeated 1000 times.
n_ten = np.random.binomial(10, 0.5, 1000)
# let us plot the same.
sns.set_style("whitegrid")
sns.countplot(x=n_ten)

# now let us see for example the probability of getting 5 heads is approx 250 times out of 1000 times i.e. 0.25, similarly we can see for others as well.
```

```python
# now let us take a practical example.

# A manufacturing process has 12 % error rate during production. The customer decides to check 20 random samples & will only accept if there are 2 or less errors defectives.
# What is the probability of getting accepted i.e. getting 2 or less defectives? (here n=20, k=2, p=0.12 )
p2orless = binom.cdf(2, 20, 0.12)  # cdf - cumulative density function.
print(f"Probability of getting 2 or less defectives is {p2orless}.")

# now in order to calculate the probability of getting exactly 2 defectives.
p2exact = binom.pmf(2, 20, 0.12)   # pmf - probability mass function.
print(f"Probability of getting exactly 2 defectives is {p2exact}.")

# also let us get the probability of getting more than 2 defectives.
p2ormore = binom.sf(2, 20, 0.12)   # sf - survival function which is same as (1 - binom.cdf()).
print(f"Probability of getting more than 2 defectives is {p2ormore}.")
```

```python
# now let us plot the probabilities of getting exactly 0 to 20 defectives i.e. pmf via barplot.
x_ax = np.arange(0, 21)
y_ax = binom.pmf(x_ax, 20, 0.12)
plt.title("PMF PLot")
sns.barplot(x=x_ax, y=y_ax)
```

```python
# also let us plot the cumulative probabilities i.e. cdf via barplot.
x_ax = np.arange(0, 21)
y_ax = binom.cdf(x_ax, 20, 0.12)
plt.title("CDF PLot")
sns.barplot(x=x_ax, y=y_ax)
```

```python
# let us now calculate the mean, variance & standard deviation of a Binomial Distribution.
binom_mean = binom.mean(20, 0.12)
binom_variance = binom.var(20, 0.12)
binom_std = binom.std(20, 0.12)
print(binom_mean)
print(binom_variance)
print(binom_std)
```

##### Poisson Distribution
- Poisson Distribution is similar to Binomial Distribution in the fact that it is also a discrete distribution & measures the number of successes.
- However unlike Bimonial Distribution here the possibilities of success are infinite.
- One classical example is the people standing in a queue, here number of people can be 0, 10, 1000, 10000 etc.
- Another example could be the number of road accidents happening every hour in a city. It could be 0, 10, 20, 100, 1000 etc, thus the possibilities are endless.
- A Poisson Experiment also result in success or failure, however in this case the average number of successes that occurs in a specified region is known.
- Outcomes are also random in this case & the occurrence of one outcome does not influence the chance of another outcome.
- However the outcomes are rare relative to the possible outcomes eg : actual no of accidents occurring every hour are very low relative to the assumed value.
- Poisson Probability is the probability p(x; μ) that exactly x successes occur in an experiment where the mean number of successes is μ.
- Formula : **P(x, μ) = e^(-μ) . μ^x/x!**
- In a Poisson Distribution, mean & variance are equal.

```python
# let us first import the relevant library.
from scipy.stats import poisson

# let us now simulate a poisson experiment, where we will generate an array of 1000 numbers with mean as 3.6.
my_data = np.random.poisson(3.6, 1000)

# alo let us visualize the distribution via countplot.
sns.countplot(x=my_data)
```

```python
# now let us see through a practical example.

# In an medical billing process, on an average 3.6 invoices are processed every 10 minutes. What is the probability of processing:-
# exactly 7 invoices in 10 minutes. (here k=7, μ=3.6)
p_exactly7 = poisson.pmf(7, 3.6)
print(p_exactly7)

# 7 invoices or less than 7 in 10 minutes.
p_7orless = poisson.cdf(7, 3.6)
print(p_7orless)

# more than 7 invoices in 10 minutes.
p_7ormore = poisson.sf(7, 3.6)
print(p_7ormore)
```

```python
# let us now plot the probabilities against different values i.e. pmf via barplot.
x_ax = np.arange(0, 11)
y_ax = poisson.pmf(x_ax, 3.6)
plt.title("PMF PLot")
sns.barplot(x=x_ax, y=y_ax)
```

```python
# also let us plot the cumulative probabilities i.e. cdf via barplot.
x_ax = np.arange(0, 11)
y_ax = poisson.cdf(x_ax, 3.6)
plt.title("CDF PLot")
sns.barplot(x=x_ax, y=y_ax)
```

##### Normal Distribution
- Normal Distribution is a continuous distribution i.e pertaining to continuous data.
- It is symmetrically distributed & it resembles a bell shaoped curve with long tails.
- Mean / Mode & Median are same.
- The shape of the normal distribution is defined by two factors i.e. **Mean & Standard Deviation**.
- About 68% of the area under the curve falls within 1 std deviation of the mean.
- About 95% of the area under the curve falls within 2 std deviations of the mean.
- About 99.7% of the area under the curve falls within 3 std deviations of the mean.
- The total area under the normal curve = 1.00 & the probability of any particular value is = 0.
- The probability that X is greater than or less than a value = area under the normal curve in that direction (i.e. less to the left & greater to the right of that value).
- Normal Probability **P(x) = e^(-(x-μ)^2 / 2.σ^2) / σ.sqrt(2π)**, where x is a normal random variable, μ is the mean & σ is the standard deviation.
- We will not be using the above formula as the probability for any particular value is 0, here comes the z value & standard normal distribution concepts.
- Z value / Standard Score, **z = (x - μ) / σ**, where x is the value of the element, μ is the population mean & σ is the population standard deviation.
- We standardize every single normal distribution into a standard normal distribution which is having mean of 0 & standard deviation of 1 by creating a z-value.


```python
# let us first import the relevant library.
from scipy.stats import norm

# let us first simulate a normal distribution by drawing 1 million random samples from a population having a mean of 100 & standard deviation of 2 (here μ(loc)=100, σ(scale)=2).
norm_data = np.round(np.random.normal(100, 2, 1000000), 2)

# let us now create a histogram along with the distribution plot for the above random data.
sns.histplot(x=norm_data, kde=True)
```

```python
# now let us demonstrate the below three characteristics of a normal distribution:-
# About 68% of the area under the curve falls within +/- 1 std deviation of the mean.
sum1 = sum((norm_data > 98) & (norm_data < 102))
print(sum1)  # about 681843 out of 1000000 points fall within +/- 1 std deviation of the mean which approximates to 68%.

# About 95% of the area under the curve falls within +/- 2 std deviations of the mean.
sum2 = sum((norm_data > 96) & (norm_data < 104))
print(sum2)  # about 954532 out of 1000000 points fall within +/- 1 std deviation of the mean which approximates to 95%.

# About 99.7% of the area under the curve falls within +/- 3 std deviations of the mean.
sum3 = sum((norm_data > 94) & (norm_data < 106))
print(sum3)  # about 997334 out of 1000000 points fall within +/- 1 std deviation of the mean which approximates to 99.7%.
```

```python
# now let us consider a practical example.
# In a medical billing process, it takes on an average 150 min to process an invoice with a standard deviation of 2 min. What percentage of invoices will be processed:-
# in more than 153 min. (here x=153, μ=100, σ=2)
p_more_than_153 = norm.sf(153, 150, 2)
print(p_more_than_153) 

# between 148 & 152 min.
p_more_than_148 = norm.sf(148, 150, 2)
print(p_more_than_148) 
p_more_than_152 = norm.sf(152, 150, 2)
print(p_more_than_152) 

p_betw_148_152 = p_more_than_148 - p_more_than_152
print(p_betw_148_152) 
```

```python
# let us now create the distribution plot for the above data via lineplot.
x_data = np.linspace(144, 156, 1000)
y_data = norm.pdf(x_data, 150, 2)
sns.lineplot(x=x_data, y=y_data)

# let us now add a vertical line to the plot.
plt.axvline(153, color="red")
```

```python
# let us also calculate other relevant measures for a normal distribution.

# z-value pertaining to 153 min.(here x=153, μ=150, σ=2) 
z = (153 - 150) / 2
print("Z-Value:", z)

# to check the probability of values to the left of z=1.5.
prob_l = norm.cdf(z)
print(prob_l)

# in order to again get a z-value from the probability value.
val = norm.ppf(prob_l)  # ppf - percent point function.
print(val)

# to check the probability of values to the right of z=1.5.
prob_r = norm.sf(z)
print(prob_r)

# in order to again get a z-value from the probability value to the right of z=1.5.
inv_val = norm.isf(prob_r)
print(inv_val)
```

##### Sampling
- Sampling is a part of inferential where we infer about the population based on samples.
- In sampling we draw the required number of samples from a population & then perform statistical estimations on the sample
- Based on the estimations done on the sample we conclude about the characteristics of the population.
- The decsriptive measures like mean & standard deviation for the entire population is termed as a **Parameter** & that for the sample is termed as **Statistic**.
- Thus for population we have **Population Mean(μ) & Population Standard Deviation(σ)** & for sample we have **Sample mean (x̄) & Sample Standard Deviation(s)**.
- So there will be a **Point Estimate** as well an **Interval Estimate** basis which we can infer about the population.
- Eg : In a medical billing process we have an average(mean) time for processing an invoice to be 100 mins & standard deviation to be 1 min.
- now let us draw 4 random invoices from this population of invoices & record the time for their processing which is let's say 102, 101, 99, 100.
- Eg : Now the time for processing an invoice in the medical billing process comes out to be 100.5 +/- 1 which means it takes approx 99.5 min to 101.5 min to process an invoice.
- Here 100.5 is the average processing time which is calculated basis the random invoices that are drawn from the population.
- This estimated range from 99.5 min to 101.5 min is known as **Confidence Interval(CI)**.
- The level of confidence generally taken to be 95%.
- Formula : **CI = x̄ +/- z(𝛼/2).σ/sqrt(n)** where z(𝛼/2).σ/sqrt(n) refers to the **Margin of Error**.

```python
# let us take the example where in a medical billing process we have an average(μ) time for processing an invoice to be 100 mins & standard deviation(σ) to be 1 min.
# now let us draw 4(n=4) random invoices from this population of invoices & record the time for their processing which is let's say 102, 101, 99, 100.

# let us first calculate the Sample mean(x̄).
import statistics
sample_data = [102, 101, 99, 100]
sample_mean = statistics.mean(sample_data)
print("Sample mean(x̄):", sample_mean)

# now let us calculate the Z-value corresponding to 95% confidence interval or 𝛼 = 0.05.
z_val = norm.isf(0.025)
print("Z-value for 0.025(𝛼/2):", z_val)

# now let us calculate the Margin of Error.
margin_of_error = (z_val * 1) / np.sqrt(4) 
print("Margin of Error:", margin_of_error)

# finally let us calculate the Confidence Interval.
CI_upper = sample_mean + margin_of_error
CI_lower = sample_mean - margin_of_error
print("Upper Bound:", CI_upper)
print("Upper Bound:", CI_lower)

# conclusion : Thus we can say that with 95% confidence we can say that the average invoice processing time is approximately between 99.5 min to 101.5 min.
```

##### Central Limit Theorem (CLT)
- There are two key aspects in Central LImit Theorem.
- One is that for almost all populations, the sampling distribution of the means(sample means) can be approximated closely by a normal distribution, provided the sample size is sufficiently large.
- If the variable has a mean of μ & variance of σ^2, as the sample size n increases, the sample mean approaches a normal distribution with mean μx̄ & variance σ^2x̄.
- Thus **μx̄ = μ** & **σx̄ = σx/sqrt(n)**, here σx̄ is known as the **Standard Error of the Mean** & it is the standard deviation of the sampling distribution of the sample mean.

```python
# let us simulate the rolling of a die.

# let us first roll the die 100 times.
one_hundred = np.random.randint(1, 7, 100)
print(one_hundred)

# let us see the distribution through a histogram.
sns.histplot(x=one_hundred, bins=6)
```

```python
# let us now roll the dice 1 million times.
one_million = np.random.randint(1, 7, 1000000)

# let us see this distribution through a histogram.
sns.histplot(x=one_million, bins=6)

# let us also find the mean & standard deviation of this distribution.
print(np.mean(one_million))
print(np.std(one_million))
```

```python
# now let us draw four samples from this distribution.
one_million = one_million.reshape(250000, 4)

# now we will calculate the mean value for each row.
mean_data = one_million.mean(axis=1)

sns.histplot(x=mean_data, bins=6, kde=True)
```

```python
# let us also calculate the mean & standard deviation for the mean data.
sample_mean = np.mean(mean_data)
print(sample_mean)  # here sample mean is 3.5 which is the same as the population mean previously calculated.

sample_std = np.std(mean_data)
print(sample_std)  # here sample standarrd deviation is 0.85 which is half of the population standard deviation.
```

```python
# let us say now instead of 4 we draw 100 samples from the population.
mean_100_samples = one_million.reshape(10000, 100).mean(axis=1)
print(mean_100_samples)

# now let us plot the histogram.
sns.histplot(x=mean_100_samples, bins=6, kde=True)  # this plot very much resembles the normal distribution.
```

```python
# let us also calculate the mean & standard deviation for the mean data.
sample_mean = np.mean(mean_100_samples)
print(sample_mean)  # here sample mean is 3.5 which is the same as the population mean previously calculated.

sample_std = np.std(mean_100_samples)
print(sample_std)  # here sample standarrd deviation is 0.17 which is one-tenth of the population standard deviation.
```
