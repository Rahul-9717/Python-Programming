##### **Matplotlib**
- It is the primary library for plotting & visualization & many other visualization libraries are built upon it like seaborn & pandas built-in vizualization.
- We have the two different approaches in Matplotlib : *Functional method* which covers the Matplotlib basics & *OOP method* which covers Matplotlib figures & subplots.

```python
# first of all let us install the matplotlib library.
#%pip install matplotlib
```

```python
# import matplotlib library.
import matplotlib

# check the version.
matplotlib.__version__
```

```python
# sinc most of the matplotlib utilities are present in the pyplot module, so let us import the same.
import matplotlib.pyplot as plt
```

##### Let us first see the basic functional method of plotting in Matplotlib.

```python
# now let us create a very basic plot from an array.
import numpy as np
x = np.arange(0, 10)
y = np.arange(0, 20, 2)
plt.plot(x, y)  # this returns a matplotlib 2d line object along with the visualization.

# Note : in case the plot does not show up add plt.show() at the end of your code.(mandatory to use if running a .py script)
```

```python
# if you don't want to see the matplotlib object declaration just add semicolon to the plot command.
plt.plot(x, y);
```

```python
# let us now add a title to the plot.
plt.plot(x, y)
plt.title("Linear Plot");
```

```python
# let us now add the axis labels to the plot.
plt.plot(x, y)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Linear Plot");
```

```python
# let us now adjust the values shown in the X & Y axis scales.
plt.plot(x, y)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.xlim(0, 10)
plt.ylim(0, 20)
plt.title("Linear Plot");
```

```python
# let us now save our plot as a png file.
plt.plot(x, y)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.xlim(0, 10)
plt.ylim(0, 20)
plt.title("Linear Plot")
plt.savefig('Plot.png') # in case we are saving this file in another directory then we have to specify the full path.
```

##### Now we will shift our attention to the object oriented method for visualization in Matplotlib.
- This method leverages the Figure Object which is like a blank canvas & then we will then add axes & plots to this object.

```python
# let us first create the figure object.
plt.figure() # by default this will create a blank canvas of 432 x 288 pixels.
```

```python
# now if we require to modify the size of the canvas.
plt.figure(figsize=(10,10)) # this will create a blank canvas of 10" x 10" i.e 720 x 720 pixels.
```

```python
# let us now add the axes to the figure object in order to define the scale of our canvas.
plot = plt.figure()
axes = plot.add_axes([0, 0, 1, 1]) # here the base of the figure starts at 0 & both the width & height of the fig is 1.0.
# here the figure canvas starts at coordinates (0,0) which is at bottom left & end at (1,1) which is the top right.
```

```python
# let us now add our plot to this figure canvas.
plot = plt.figure()
axes = plot.add_axes([0, 0, 0.5, 0.5])
axes.plot(x, y);
```

```python
# let us now implement the figures & axes concepts in creating our visualizations.
x = np.linspace(0, 5, 100)
y = np.linspace(0, 25, 100)
viz = plt.figure()
axes = viz.add_axes([0, 0, 1, 1])
axes.plot(x, y);
```

```python
# in order to create a new set of axes within an existing axes in the figure canvas.
a = np.arange(0, 10)
b = a ** 3
viz = plt.figure()
axes1 = viz.add_axes([0, 0, 1, 1])
axes2 = viz.add_axes([0.1, 0.6, 0.3, 0.3]) # here the values represent the offset from the x & y axes as well as height & width.
axes1.plot(a, b)
axes2.plot(a, b);
```

```python
# now let us add the X & Y labels, set their limits & also add title to each of the above plots.
viz = plt.figure()
axes1 = viz.add_axes([0, 0, 1, 1])
axes1.set_xlim(0, 10)
axes1.set_ylim(0, 1000)
axes1.set_xlabel("A")
axes1.set_ylabel("B")
axes1.set_title("AB Plot")

axes2 = viz.add_axes([0.1, 0.3, 0.3, 0.3])
axes2.set_xlim(0, 4)
axes2.set_ylim(0, 100)
axes2.set_xlabel("A")
axes2.set_ylabel("B")
axes2.set_title("AB Plot Zoomed In View")
axes1.plot(a, b)
axes2.plot(a, b);
```

```python
# we can enhance the figure size & clarity by adding the figsize & dpi parameters
viz = plt.figure(figsize=(3,2), dpi=100)
axes = viz.add_axes([0, 0, 1, 1])
axes.plot(a, b);
```

```python
# let us now save the above plot.
viz.savefig('new_plot.png') # now in the png file we are not able to see the axes limits as we are seeing in the plot above.
```

```python
# in order to see the axis limit values inside the png file.
viz.savefig("new_plot.png", bbox_inches="tight")
```

##### Let us now focus our attention to subplots
- Matplotlib enables us to create a multiple plots side by side by leveraging the plt.subplots() function.
- This will eliminate the manual effort invested in creating a figure object & then manually adding & arranging sets of axes.
- The plt.subplots()function will return a tuple containing the figure canvas & a numpy array which holds the axes objects.

```python
# let us create a subplot containing multiple axes in the overall figure canvas.
fig, axes = plt.subplots(nrows=2, ncols=2) 
axes # axes is basically the numpy array holding each of the axes basis the position in the overall canvas in this case it's a 2x2 numpy array.
```

```python
# let us now create subplots.
x = np.linspace(0, 5, 100)
y = np.linspace(0, 25, 100)
a = np.arange(0, 10)
b = a ** 3
fig, axes = plt.subplots(nrows=1, ncols=2)

# now in order to create plots in these axes.
axes[0].plot(x,y)
axes[1].plot(a,b);
```

```python
# we can also create a 2x2 fig, axes.
fig, axes = plt.subplots(nrows=2, ncols=2)

# now in order to create plots in these axes.
axes[0][0].plot(x,y)
axes[1][1].plot(a,b);
```

```python
# let us take the case on one column & multiple rows scenario.
fig, axes = plt.subplots(nrows=3, ncols=1)

# now in order to create plots in these axes.
axes[0].plot(x,y)
axes[1].plot(a,b)
axes[2].plot(x,y);
```

```python
# we can also leverage for loop to create subplots for cases involving vertically stacked subplots.
fig, axes = plt.subplots(nrows=3, ncols=1)
for viz in axes:
    viz.plot(a, b)
```

```python
# in order to have a cleaner layout with proper spacing.
fig, axes = plt.subplots(nrows=2, ncols=2)
axes[0][0].plot(x, y)
axes[1][1].plot(x, y)
axes[0][1].plot(a, b)
axes[1][0].plot(a, b);

plt.tight_layout()
```

```python
# in order to manually adjust the layout spacing.
fig, axes = plt.subplots(nrows=2, ncols=2)
axes[0][0].plot(x, y)
axes[1][1].plot(x, y)
axes[0][1].plot(a, b)
axes[1][0].plot(a, b);

fig.subplots_adjust(wspace=0.5, hspace=0.5)
```

```python
# in order to set the axes labels as well title for any of these subplots.
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 5), dpi=100)
axes[0][0].plot(x, y)
axes[0][0].set_xlim(0,5)
axes[0][0].set_ylim(0,30)
axes[0][0].set_xlabel("X")
axes[0][0].set_ylabel("Y")
axes[0][0].set_title("XY Plot")

axes[1][1].plot(x, y)

axes[0][1].plot(a, b)
axes[0][1].set_xlim(0,10)
axes[0][1].set_ylim(0,1000)
axes[0][1].set_xlabel("A")
axes[0][1].set_ylabel("B")
axes[0][1].set_title("AB Plot")

axes[1][0].plot(a, b);

# we can also have an overall title for the entire subplot grid.
fig.suptitle("Plot Grid", fontsize=15)

plt.tight_layout()
```

##### Let us now focus on the styling aspects for our plots.

```python
# let us add a legend to our plot.
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y, label="XY")
axes.plot(a, b, label="AB")
axes.legend();
```

```python
# we can also choose to locate the legend in our plot.
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y, label="XY")
axes.plot(a, b, label="AB")
axes.legend(loc="lower right");
```

```python
# in order to set the legend location based on coordinates.
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y, label="XY")
axes.plot(a, b, label="AB")
axes.legend(loc=(1.01,0.5)); # here we have passed a tuple containing X & y coordinates.
```

```python
# let us now assign a specific line color.
x = np.array([3, 6, 8, 9, 12, 15, 20, 17, 5, 1])
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, color="#eb3462"); # here we have choosen the hex code from hex color picker in order to assign a customized color. 
```

```python
# let us now set the line width.
x = np.array([3, 6, 8, 9, 12, 15, 20, 17, 5, 1])
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, color="#eb3462", lw=3) # this will create the plot having line width 3 times the default width.
```

```python
# let us now set the line style.
x = np.array([3, 6, 8, 9, 12, 15, 20, 17, 5, 1])
y = x ** 0.5
z = x ** 1.25
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, color="#eb3462", lw=5, ls="--")
ax.plot(y, color="green", lw=4, ls="-.")
ax.plot(z, color="blue", lw=3, ls=":")
```

```python
# we can also create a custom line style.
x = np.array([3, 6, 8, 9, 12, 15, 20, 17, 5, 1])
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
viz = ax.plot(x, color="purple", lw=3)
type(viz) # here viz is basically a list object.
viz[0].set_dashes([2, 2, 10, 2]) # [no of dashes, no of spaces & so on..]
```

```python
# let us now add markers to our line & style them.
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
y = x ** 0.5
z = x ** 1.25
ax.plot(x, color="purple", lw=2, marker="o", ms=12)
ax.plot(y, color="green", lw=2, marker="s", ms=10)
ax.plot(z, color="blue", lw=2, marker=".", ls="--", ms=15)
```
